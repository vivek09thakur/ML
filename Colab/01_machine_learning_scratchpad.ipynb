### **VISUALIZING BOTH DATASETS**

```
x1,y1 = TRAIN_DATASET_1['median_house_value'], TRAIN_DATASET_1['median_income']
```

```
train_x,train_y = np.array(x1),np.array(y1)
```

```
plt.hist(train_x)
plt.show()
```

```
train_x,train_y = train_x[:30],train_y[0:30]
```

```
plt.hist(train_x)
```

```
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(train_x, train_y)
```

```
def model(x):
  return slope * x + intercept

mymodel = list(map(model, train_x))
```

```
plt.scatter(train_x,train_y)
plt.plot(train_x, mymodel)
plt.show()
```
